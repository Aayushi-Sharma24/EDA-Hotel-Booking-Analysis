#Let's look into cancellation rate among different type of hotel.

```python
# Enlarging Figure size
plt.figure(figsize = (10,8))
# Function to plot % over the graph
def with_hue(plot, feature, Number_of_categories, hue_categories):
    a = [p.get_height() for p in plot.patches]
    patch = [p for p in plot.patches]
    for i in range(Number_of_categories):
        total = feature.value_counts().values[i]
        for j in range(hue_categories):
            percentage = '{:.2f}%'.format(100 * a[(j*Number_of_categories + i)]/total)
            x = patch[(j*Number_of_categories + i)].get_x() + patch[(j*Number_of_categories + i)].get_width() / 2 - 0.15
            y = patch[(j*Number_of_categories + i)].get_y() + patch[(j*Number_of_categories + i)].get_height() 
            ax.annotate(percentage, (x, y), size = 12,verticalalignment='bottom')
    plt.show()

# Plotting the data using Countplot
ax = sns.countplot(x= 'is_canceled',data = df,hue = 'hotel',palette= 'Set1')
plt.title("Cancelation Status By Hotel Type ", fontsize=15)
# Setting X and Y labels
plt.xlabel("Cancelation Status")
plt.ylabel(" Reservations Count")
with_hue(ax,df.is_canceled,2,2)
```

```python
# Enlarging Figure size
plt.figure(figsize = (10,8))
# Function to plot % over the graph
def with_hue(plot, feature, Number_of_categories, hue_categories):
    a = [p.get_height() for p in plot.patches]
    patch = [p for p in plot.patches]
    for i in range(Number_of_categories):
        total = feature.value_counts().values[i]
        for j in range(hue_categories):
            percentage = '{:.2f}%'.format(100 * a[(j*Number_of_categories + i)]/total)
            x = patch[(j*Number_of_categories + i)].get_x() + patch[(j*Number_of_categories + i)].get_width() / 2 - 0.15
            y = patch[(j*Number_of_categories + i)].get_y() + patch[(j*Number_of_categories + i)].get_height() 
            ax.annotate(percentage, (x, y), size = 12,verticalalignment='bottom')
    plt.show()

# Plotting graph using countplot
ax = sns.countplot(x= 'is_canceled',hue = 'is_repeated_guest',data = df,palette= 'Greens')
# Setting Title of graph
plt.title("Effect of repeated guest on Cancellations", fontsize=15)
with_hue(ax,df.is_canceled,2,2)
# Show object
plt.show()
```

####From the above graph, we can depict that most of the repeated guests do not cancel their reservations and its obvious

```python
# Enlarging figure size
plt.figure(figsize=(10,8))
# Checking if Assigned room is same as Reserved room
df['same_room_assigned'] = np.where(df.reserved_room_type==df.assigned_room_type, 1 , 0)
# Function to change the width of bars in barplot
def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)

# Plotting graph using barplot
ax = sns.barplot(x='same_room_assigned', y = 'is_canceled', data=df)
change_width(ax, .35)
# Calculating percentage of same room assigned 
percentage = df.groupby('same_room_assigned')['is_canceled'].mean().to_list()
# Annotate the percentage on graph
for i in range(0,2):
  plt.text(i-0.1,round(percentage[i],2)-0.03,str(round(percentage[i]*100,2))+'%',fontsize=14,horizontalalignment = 'left')

# Setting title and XY labels
plt.title('Room assigned vs Cancellation rate',fontsize=15)
plt.xlabel('Room assigned')
plt.ylabel('Cancellation rate')
plt.legend(['Same room not assigned'])
plt.show()
```
